# üîπ Construindo um Sistema Banc√°rio em Python üè¶üí∞

## üìå Descri√ß√£o
Voc√™ foi contratado para desenvolver um sistema banc√°rio simples utilizando Python! O sistema deve permitir que o usu√°rio realize dep√≥sitos, saques, visualize o extrato e encerre o programa. Todas as opera√ß√µes devem ser registradas e exibidas corretamente.

## üìå Objetivo:
Criar um sistema funcional que simule opera√ß√µes banc√°rias b√°sicas.

## üìå Regras do Sistema
- 1Ô∏è‚É£ Dep√≥sito

O usu√°rio pode inserir um valor positivo e adicion√°-lo ao saldo da conta.
A opera√ß√£o deve ser registrada no extrato.

- 2Ô∏è‚É£ Saque

O usu√°rio pode retirar um valor do saldo da conta.
A opera√ß√£o deve ser registrada no extrato.

- 3Ô∏è‚É£ Extrato

Exibe todas as transa√ß√µes feitas (dep√≥sitos e saques).
Exibe o saldo atual da conta.

- 4Ô∏è‚É£ Sair

O sistema deve permitir que o usu√°rio encerre o programa digitando a op√ß√£o correspondente.

- 5Ô∏è‚É£ Valida√ß√µes (B√¥nus)

O sistema n√£o pode permitir saques se o saldo for insuficiente.
O usu√°rio n√£o pode depositar valores negativos.

## üìå Entrada
O usu√°rio interage com um menu num√©rico, escolhendo op√ß√µes para realizar opera√ß√µes banc√°rias.

## üìå Sa√≠da
Mensagens informando o status das opera√ß√µes e um extrato detalhado das transa√ß√µes.

## üìå Exemplo de Uso
Exemplo de intera√ß√£o do usu√°rio:

```
========== BANCO PYTHON ==========
[1] Dep√≥sito
[2] Saque
[3] Extrato
[4] Sair
Escolha uma op√ß√£o: 1
Informe o valor do dep√≥sito: R$ 500.00
Dep√≥sito de R$ 500.00 realizado com sucesso!

========== BANCO PYTHON ==========
[1] Dep√≥sito
[2] Saque
[3] Extrato
[4] Sair
Escolha uma op√ß√£o: 2
Informe o valor do saque: R$ 200.00
Saque de R$ 200.00 realizado com sucesso!

========== BANCO PYTHON ==========
[1] Dep√≥sito
[2] Saque
[3] Extrato
[4] Sair
Escolha uma op√ß√£o: 3

========== EXTRATO ==========
Dep√≥sito: +R$ 500.00
Saque: -R$ 200.00

Saldo atual: R$ 300.00
=============================

========== BANCO PYTHON ==========
[1] Dep√≥sito
[2] Saque
[3] Extrato
[4] Sair
Escolha uma op√ß√£o: 4
Saindo do sistema...
```

````python
valor = 0.0
valor_real = 0.0
extrato = []
def mostrar_menu():
    print("========== BANCO PYTHON ==========")
    print("[1] Dep√≥sito")
    print("[2] Saque")
    print("[3] Extrato")
    print("[4] Sair")
while True:
    mostrar_menu()
    try:
      escolha = int(input("Escolha uma op√ß√£o: "))
      if escolha == 1: #Op√ß√£o de Dep√≥sito
          valor = float(input("Informe o valor do dep√≥sito: R$ "))
          if valor > 0.0:
            valor_real = valor_real + valor
            extrato.append(f"Dep√≥sito: +R$ {valor:.2f}")
            print(f"Dep√≥sito de R$ {valor} realizado com sucesso!\n")
          else:
            print("N√£o pode depositar Valor Negativo\n")
      elif escolha == 2: # Op√ß√£o de Saque
            valor = float(input("Informe o valor do saque: R$ "))
            if valor > 0:
                if valor_real >= valor:
                    valor_real -= valor
                    extrato.append(f"Saque: -R$ {valor:.2f}")
                    print(f"Saque de R$ {valor:.2f} realizado com sucesso!\n")
                else:
                    print("N√£o tem saldo suficiente para realizar o saque.\n")
            else:
                 print("N√£o pode Sacar Valor Negativo ou Zero\n")
      elif escolha == 3: # Op√ß√£o Extrato
            print("\n========== EXTRATO ==========")

            if extrato: # Verifica se a lista n√£o est√° vazia
                for transacao in extrato: # Para cada item na lista extrato...
                    print(transacao)     # ...imprima esse item.
            else:
                print("N√£o foram realizadas movimenta√ß√µes.")
            print(f"\nSaldo atual: R$ {valor_real:.2f}")
            print("=============================\n")
      elif escolha == 4:
          print("Saindo do sistema...")
          break
      else:
        print("Op√ß√£o Inv√°lida, tente novamente \n")
    except ValueError:
        print("Op√ß√£o Inv√°lida, tente novamente \n")
````

#üîπ Simulador de Caixa Eletr√¥nico üí∞

````python
saldo = 1000
valorSaldo = 0.0
def menu():
  print("MENU:\n1 - Depositar\n2 - Sacar\n3 - Sair")
while True:
  menu()
  try:
    escolha = int(input("Escolha uma op√ß√£o: "))
    if escolha == 1:
      valorSaldo = float(input("Digite o valor do dep√≥sito: "))
      if valorSaldo < 0:
        print("Valor Inv√°lido")
      else:
        saldo = saldo + valorSaldo
      print(f"Dep√≥sito realizado! Novo saldo: R$ {saldo}\n")
    if escolha == 2:
      valorSaldo = float(input("Digite o valor do Saque: "))
      saldo = saldo - valorSaldo
      if saldo < 0:
        saldo = abs(saldo)
        saldo = valorSaldo - saldo
        print("Saldo insuficiente\n")
      else:
        print(f"Saque realizado! Novo saldo: R$ {saldo}")
    if escolha == 3:
      print("Obrigado por usar nosso caixa eletr√¥nico!")
      break
    elif escolha <= 0 or escolha >3:
      print("Op√ß√£o Inv√°lida\n")
  except ValueError:
        print("Op√ß√£o Inv√°lida, tente novamente \n")
````
